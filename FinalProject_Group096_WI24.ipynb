## Background and Prior Work

Introduction:
	Airbnb is an online marketplace which manages short-term homestays through homeowners. Users can scroll through listings and rent out a home to stay at for a certain period of time. A cut of the money goes to the company as commission. As traveling becomes more popular through social media and remote jobs, Airbnb also becomes more prevalent and widespread among the general public. It is estimated that every second, six guests check into a listing (Woodward, AIRBNB STATISTICS [2024]: USER & MARKET GROWTH DATA)<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). With the popularity of Airbnb also comes ethical dilemmas, such as the issue of Airbnbs driving locals out of their homes through gentrification. By observing datasets about Airbnb listing prices, we can make inferences about the future prices of current Airbnb listings and any potential changes in them. This knowledge can provide hosts and users a more transparent and efficient experience, as well as ensure hosts are not price gouging their listings.

Prior work:
- Previous work that analyzed if past data can predict changes in Airnbnb prices showed different results, based on the different array of listing factors that were integrated to help predict the listing price. One study claims that the 3 most important factors that can help predict the listing price are the number of people accommodated, bedrooms, and beds, all of which have a Pearson's Correlation Factor of around 0.45 with the Listing Price (Karkala, Predictive Price Modeling for Airbnb listings)<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). Another study found that the most important factors in predicting price are the type of listing (entire home/room), amount of people accommodated, the type of property, and the amount of bathrooms (Carrillo, Predicting Airbnb prices with machine learning and location data)<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). Overall, most studies found a strong correlation between price increases and the previous renters’ rating of location. Results differed also due to the different data cleaning and exploration methods utilized, such as EDA (Exploratory Data Analysis) in Carrillo’s study, which revealed seasonality in listings and how they influenced price changes due to local events exclusive to certain locations, such as the Edinburgh Fringe Festival. 

Expanding on Prior work:
- In how we would expand on prior work, we would use some of the variables that Karkala used but also include more variables to see if they have a bigger effect in predicting prices for Airbnb. We also would look at neighborhoods in New York City as some areas in New York can be much pricier than others. Lastly, we are interested in whether having transit would have a big impact on prices as New York is run primarily on public transportation, so having a close bus stop way stop can have a big impact on price. We would also try to see if having a good host profile (like response rate and rating) would make the model make the predictor more accurate. The overall objective of our research project is to lower the Mean Squared Error of predicting Airbnb price listings in order to get a more accurate predictor model compared to Karkala.

## Why these variables? We feel like these variables would contribute the most because:
1. `'Property Type'`: This refers to the kind of property being listed, such as an apartment, house, condominium, loft, etc. The property type is important because different types of properties usually have different market values and appeal to different kinds of guests. For instance, a house might be more appealing to large groups or families which can lead to a higher price as there are more people.

2. `'Room Type'`: This variable includes 'Entire home/apt', 'Private room', 'Shared room. Room type directly affects the price because entire homes or apartments typically cost more than private or shared rooms, reflecting the level of privacy and space offered to guests.

3. `'Bedrooms'`: The number of bedrooms is also a useful variable in predicting a listing price because it often correlates with the accommodation's capacity to host guests. In addition, more bedrooms can accommodate more guests, which increases the price due to the fact that there is more people. We will imputate the N/A values with the median `'Bedrooms'` to maintain the integrity of our dataset

4. `'Bathrooms'`: The number of bathrooms also influences the price of the listing because bathrooms also contribute the amount of space you rent. So the more bathrooms a listing has, the more space there is in the listing, which increases the price. We will imputate the N/A values with the median `'Bathrooms'` to maintain the integrity of our dataset

5. `'Accommodates'`: This refers to the maximum number of guests a listing can accommodate. Listings that can accommodate more guests are typically priced higher due to the flexibility the listing offers. This variable is related to the number of bedrooms but provides a more exact measurement.

6. `'Neighborhood Group Cleansed'`: The neighborhood group where the listing is located can impact the price due to factors like accessibility, safety, proximity to transit, etc. For instance, listings in Manhattan might be more expensive due to the popularity among tourists.

7. `'Neighborhood Cleansed'`: Within each neighborhood group, specific neighborhoods can have vastly different characteristics and appeal, further refining the price prediction. For example, within Brooklyn, a listing in Williamsburg might be priced differently from one in Bay Ridge due to differences in popularity, accessibility, and activites.

8. `'Host Response Rate'`: We believe that the host response rate will be a good indicator for predicting Airbnb prices in New York City because the higher the host response rate, the hosts who are more responsive and engaged with their guests might be able to put their prices higher for their listings due to the higher quality of service and guest care. We will imputate the N/A values with the median `'Host Response Rate'` to maintain the integrity of our dataset

9. `'Transit'`: The accessibility of transit options from the listing plays a significant role in determining its price. Listings that are closer to major public transportation lines, such as subways, buses, or trams, are often more desirable for tourists and visitors. This convenience can lead to a higher demand for these listings, thereby increasing the price. We will be converting the transit column to a binary label of 0's and 1's with 0's being that the transit column is empty, and 1's being that there is a value in the column.

10. `'Review Scores Rating'`: High review scores are indicative of a host's commitment to providing a great experience, which can justify a higher price point for their listings. Positive reviews often reflect the guests' satisfaction with their stay, including the cleanliness, communication with the host, location, and the amenities offered. Listings with higher review scores are more likely to be booked, as they are seen as a safer bet by potential guests. Therefore, a higher review scores rating can directly impact the listing's ability to command a higher price, as it builds trust and credibility with potential renters. We will imputate the N/A values with the median `'Review Scores Rating'` to maintain the integrity of our dataset

## Linear Regression Prediction Model

For this model, we will use a simple linear regression prediction model.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

lm = LinearRegression()


X = nyc_airbnb_encoded.drop(columns='Price')
y= nyc_airbnb_encoded['Price']
#split the dateset into training and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

lm.fit(X_train,y_train)
from sklearn.metrics import mean_absolute_error
y_predicts = lm.predict(X_test)

print("""
        Mean Squared Error: {}
        R2 Score: {}
        Mean Absolute Error: {}
     """.format(
        np.sqrt(metrics.mean_squared_error(y_test, y_predicts)),
        r2_score(y_test,y_predicts) * 100,
        mean_absolute_error(y_test,y_predicts)
        ))
```

## Decision Tree Model

Here, we decided to use a Decision Tree Model in order to try to get a better Mean Squared Error score

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np


Reg_tree = DecisionTreeRegressor(criterion='friedman_mse', max_depth=3)
Reg_tree = Reg_tree.fit(X_train, y_train)

y_predicts = Reg_tree.predict(X_test)
print("""
        Mean Squared Error: {}
        R2 Score: {}
        Mean Absolute Error: {}
     """.format(
        np.sqrt(mean_squared_error(y_test, y_predicts)),
        r2_score(y_test, y_predicts) * 100,
        mean_absolute_error(y_test, y_predicts)
     ))
```

## Ridge Model
Lastly, we used a Ridge Model in order to achieve a lower mean squared error score than the previous two models.

```python
from sklearn import linear_model
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

# Create a pipeline that first standardizes the data then applies ridge regression
ridge_model = make_pipeline(StandardScaler(), linear_model.Ridge(alpha=0.01))

# Assuming X_train, X_test, y_train, y_test are already defined and split
ridge_model.fit(X_train, y_train)             
y_predicts = ridge_model.predict(X_test)

# Calculate metrics
mse = mean_squared_error(y_test, y_predicts)
r2 = r2_score(y_test, y_predicts) * 100
mae = mean_absolute_error(y_test, y_predicts)

print(f"""
        Mean Squared Error: {np.sqrt(mse)}
        R2 Score: {r2}
        Mean Absolute Error: {mae}
     """)
```
